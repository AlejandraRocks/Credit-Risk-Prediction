```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import pandas as pd

# Cargar el archivo con separador ; y saltar la primera fila
df = pd.read_csv("../data/default_of_credit_card_clients.csv", sep=";", header=1)

# Mostrar las primeras filas
df.head()

```

```python
# Información general del dataset
print("Forma del dataset:", df.shape)
df.info()
```

```python
# Revisión de valores nulos
df.isnull().sum()
```

```python
# Estadísticas descriptivas
df.describe()
```

```python
# Distribución de la variable objetivo
target_col = 'default payment next month'

sns.countplot(x=target_col, data=df)
plt.title("Distribución de Incumplimiento de Pago")
plt.xlabel("Incumplimiento de Pago (0 = No, 1 = Sí)")
plt.ylabel("Cantidad")
plt.show()

df[target_col].value_counts(normalize=True)
```

```python
# Mapa de calor de correlaciones
plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
plt.title("Mapa de Calor de Correlación")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Renombrar la variable objetivo si aún no lo hiciste
df.rename(columns={'default payment next month': 'default'}, inplace=True)

# Calcular correlaciones con la variable objetivo
correlations = df.corr()['default'].sort_values(ascending=False)


# Graficar
plt.figure(figsize=(10, 6))
correlations.drop('default').plot(kind='barh', color='skyblue')
plt.title('Correlación con la variable objetivo (default)')
plt.xlabel('Correlación')
plt.grid(True)
plt.tight_layout()
plt.gca().invert_yaxis()
plt.show()
```
