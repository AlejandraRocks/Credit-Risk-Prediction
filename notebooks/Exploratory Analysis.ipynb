```python
import pandas as pd

# load file
df = pd.read_csv("../data/default_of_credit_card_clients.csv", sep=";", header=1)

df.head()
```

```python
print("shape dataset:", df.shape) # General information about dataset
df.info()
```

```python
# Null values revision
df.isnull().sum()
```

```python

df.describe()  # Stadistics description
```

```python
# Objetive column
target_col = 'default payment next month'

sns.countplot(x=target_col, data=df)
plt.title("Distribution of Default on Payment")
plt.xlabel("Payment Default (0 = No, 1 = Yes)")
plt.ylabel("Amount")
plt.show()

df[target_col].value_counts(normalize=True)
```

```python

plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm") # Heat map of correlations
#plt.title("# Heat map of correlations")
plt.show()
```

```python
import matplotlib.pyplot as plt


df.rename(columns={'default payment next month': 'default'}, inplace=True)

correlations = df.corr()['default'].sort_values(ascending=False) # Calculate correlations with the target variable

plt.figure(figsize=(10, 6))
correlations.drop('default').plot(kind='barh', color='skyblue')
plt.title('Correlation with the target variable (default)')
plt.xlabel('Correlation')
plt.grid(True)
plt.tight_layout()
plt.gca().invert_yaxis()
plt.show()
```
